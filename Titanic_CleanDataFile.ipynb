What can we do with our missing data? What type of missing data do we have?
- Age is a descrete qunatitative value so we can do mean imputation or some other algorithm
- Cabin is a nominal Qualitative value so we can do Mode, or K means, or some other alogrithm
- Embarked is a nominal Qualitative value so we can do Mode, etc
- Lets start with Age! Lets look at a correlation chart too to see if age is dependent on anything too!

```python
means = df1["Age"].mean()
mean = int(means)
df1["Age"].fillna(mean,inplace = True)
df1['Age'] = df1['Age'].astype('int')
```

```python
df1 = df1
df1.head(15)
```

```python
#replace NaN in our Categorical columns
# only two columns with missing values that are categorical are cabin and embarked
# I will fill these with MODE Most common value seen B96 B98
#this doesnt make sense for me unless I don't understand ships correctly

modeC = df1["Cabin"].mode().iloc[0]
df1["Cabin"].fillna(modeC,inplace = True)

```

```python
modeE = df1["Embarked"].mode().iloc[0]
print(modeE)
df1["Embarked"].fillna(modeE,inplace = True)
```
